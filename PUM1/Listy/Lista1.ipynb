## Zad 5 - **2 pkt**

Napisz program, który sprawdzi czy podana liczba jest liczbą [Armstronga](https://pl.wikipedia.org/wiki/Liczby_Armstronga).

```kotlin
val number: Int = 153
println(checkArmstrong(number))
```
```verbatim
>>       true
```
Przykłady:
$$153 = 1^3 + 5^3 + 3^3 = 1 + 125 + 27$$
$$9 = 9^1$$
$$154 \neq 1^3 + 5^3 + 4^3 = 1 + 125 + 64$$
